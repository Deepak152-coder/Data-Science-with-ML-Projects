```python
import ast

df1['genres'] = df1['genres'].fillna('[]')
df1['genres'] = df1['genres'].apply(lambda x: ' '.join([item['name'] for item in (ast.literal_eval(x) if x.strip() and x != 'nan' else [])]))
```

```python
df1['tags'] = df1['genres'] + ' ' + df1['tagline'] + ' ' + df1['overview']
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
import re
```

```python
stopwords = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocessing(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
  word = text.split()
  word = [word for word in word if word not in stopwords]
  word = [lemmatizer.lemmatize(word) for word in word]
  text = ' '.join(word)
  return text
```

```python
df1['tags'][0]
```

```python
df1['tags'] = df1['tags'].apply(preprocessing)
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
```

```python
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,stop_words='english',ngram_range=(1,2))
```

```python
tfidf_matrix = tfidf.fit_transform(df1['tags'])
```

```python
tfidf_matrix.shape
```

```python
print(tfidf_matrix.toarray())
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
def recomend(title,n = 10):
  if title not in indices:
    return ['Movie not found']
  idx = indices[title]
  sim_score = linear_kernel(tfidf_matrix[idx],tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]

  return df['title'].iloc[similar_idx]
```

```python
recomend('Waiting to Exhale')
```
