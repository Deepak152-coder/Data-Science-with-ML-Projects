```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X,y = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=1,
    random_state=42
)
```

```python
X.shape
```

```python
X
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(X)
```

```python
cluseter = kmeans.predict(X)
centers = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],c=cluseter,cmap='rainbow')
plt.scatter(centers[:,0],centers[:,1],color='black')
plt.show()
```

```python
inetria = []
k_values = range(1,11)
for k in k_values:
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X)
    inetria.append(kmeans.inertia_)
```

```python
plt.plot(k_values,inetria,'-o')
plt.show()
```
